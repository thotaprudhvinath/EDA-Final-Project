### Step - 5 - Research Questions
Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.
Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)


```python
# Define the required specializations and job designations
required_specializations = ['computer science & engineering', 'computer engineering', 
                            'computer science and technology', 'computer science', 
                            'computer and communication engineering', 'computer networking']
required_designations = ['programmer analyst', 'software engineer', 'hardware engineer', 'associate engineer']

# Define the reference date as Jan 18, 2019
reference_date = datetime(2019, 1, 18)

# Filter the dataset based on the required specializations and job designations
# and calculate years since graduation
filtered_data = df[(df['Specialization'].isin(required_specializations)) & 
                   (df['Designation'].isin(required_designations))]
filtered_data['YearsSinceGraduation'] = reference_date.year - filtered_data['GraduationYear']

# Filter for fresh graduates (e.g., within 1 year of graduation)
years_since_graduation = 1
fresh_graduates_data = filtered_data[filtered_data['YearsSinceGraduation'] <= years_since_graduation]

# Calculate summary statistics of salary for the filtered data
salary_stats = fresh_graduates_data['Salary'].describe()

# Print summary statistics and compare with expected salary
print("Summary Statistics of Salary for Fresh Computer Science Engineers with Desired Job Designations:")
print(salary_stats)

claimed_range = (2.5, 3)  # Lakhs
if salary_stats['min'] >= claimed_range[0] and salary_stats['max'] <= claimed_range[1]:
    print("The claim is supported by the data for fresh graduates.")
    print("Reasons:")
    print("- The minimum and maximum salaries for fresh graduates fall within the claimed range.")
    print("- The majority of salaries for fresh graduates are within the claimed range.")
else:
    print("The claim is not supported by the data for fresh graduates.")
    print("Reasons:")
    print("- The minimum or maximum salary for fresh graduates falls outside the claimed range.")
    print("- A significant portion of salaries for fresh graduates are outside the claimed range.")

```
